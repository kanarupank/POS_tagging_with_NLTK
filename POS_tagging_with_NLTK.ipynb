```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize 
import json

#nltk.download('averaged_perceptron_tagger')
#nltk.download()
#nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('sentence_polarity')
 

with open('feedback_cs2012_1.txt', 'r') as myfile:
 untaggedText = myfile.read()

#print(documentText)

text = nltk.word_tokenize(untaggedText)

#perceptron 
tagged_result=nltk.pos_tag(text)
print(tagged_result)

with open('perceptron_tagged_feedback_cs2012_1.txt','w') as f:
        f.write('\n'.join('%s %s' % x for x in tagged_result))

```
